# **Testing Facebook Prophet**

```python
#Creating a new dataframe with the columns Date and Closing Price and renaming them ds and y for the model
df = df.rename(columns={'Date':'ds', 'Close':'y'})

#Removing all the other columns and only leaving ds and y
df = df[['ds', 'y']]

df
```

```python
#Get the last 30 rows of data and store them into a variable
#I am doing this to see how well this model is at predicting a future price

last = df[len(df)-30:]
last
```

```python
#Creating a new dataframe with all of the data excluding the last 20 rows of data so we can make the future predictions and check how accurate it is

df = df[:-20]
df
```

```python
#Creating the object(model)

fbp = Prophet(daily_seasonality=True)

#Fit/Train the object(model)
fbp.fit(df)
future = fbp.make_future_dataframe(periods=30)
forecast = fbp.predict(future)
```

```python
#Import plotting libraries
from fbprophet.plot import plot_plotly
plot_plotly(fbp, forecast)
```

```python
#Show the models prediction for March 20th 2022 for the price of Bitcoin

forecast[forecast.ds == '2022-03-20']['yhat']
```
